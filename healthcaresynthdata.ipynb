```python

import pandas as pd       # for working with data tables (like Excel in Python)
import numpy as np        # for math and random numbers
from datetime import datetime, timedelta  # for working with dates
import random             # for generating random choices
```

```python
random.seed(42)
np.random.seed(42)
```

```python
# How many appointments we want to simulate
N = 3000

# The 5 departments in our clinic
departments = ["Cardiology", "Orthopedics", "General Practice", "Pediatrics", "Dermatology"]

# Each department has its own doctors
providers = {
    "Cardiology":       ["Dr. Smith",   "Dr. Patel"],
    "Orthopedics":      ["Dr. Johnson", "Dr. Lee"],
    "General Practice": ["Dr. Brown",   "Dr. Garcia", "Dr. Kim"],
    "Pediatrics":       ["Dr. Wilson",  "Dr. Chen"],
    "Dermatology":      ["Dr. Davis",   "Dr. Nguyen"],
}

# Types of appointments
appt_types = ["New Patient", "Follow-Up", "Annual Checkup", "Urgent Care"]

# Possible appointment outcomes — Completed appears 3x so it's more likely
statuses = ["Completed", "No-Show", "Cancelled", "Completed", "Completed"]

# Possible claim outcomes — Paid appears 3x so it's most common
claim_status = ["Paid", "Pending", "Denied", "Paid", "Paid"]

# Clinic start date
start_date = datetime(2024, 1, 1)

print("✅ Clinic structure defined!")
print(f"Departments: {departments}")
```

```python
# This is an empty list — we'll add one row per appointment
rows = []

for i in range(1, N + 1):  # loops from 1 to 3000

    # Randomly pick a department
    dept = random.choice(departments)

    # Randomly pick a doctor FROM that department
    provider = random.choice(providers[dept])

    # Random date somewhere in 2024
    appt_date = start_date + timedelta(days=random.randint(0, 364))

    # Random scheduled hour between 8am and 5pm
    sched_time = random.randint(8, 17)

    # ⭐ THIS IS YOUR KEY FINDING
    # Cardiology & Orthopedics have higher wait times (avg 35 mins)
    # Everyone else is around 20 mins
    # np.random.normal(average, variation) generates realistic numbers
    if dept in ["Cardiology", "Orthopedics"]:
        wait_time = round(np.random.normal(35, 10))
    else:
        wait_time = round(np.random.normal(20, 8))

    # Wait time can't be negative
    wait_time = max(0, wait_time)

    # Random appointment outcome
    status = random.choice(statuses)

    # Random appointment type
    appt_type = random.choice(appt_types)

    # Random billing amount between $100 and $800
    billing_amt = round(random.uniform(100, 800), 2)

    # Claim turnaround only exists if appointment was completed
    claim_days = round(np.random.normal(18, 6)) if status == "Completed" else None
    claim_days = max(5, claim_days) if claim_days else None

    # Claim status only exists if appointment was completed
    claim_stat = random.choice(claim_status) if status == "Completed" else "N/A"

    # Slot duration depends on appointment type
    slot_mins = 30 if appt_type in ["Follow-Up", "Annual Checkup"] else 45

    # Add this appointment as one row to our list
    rows.append({
        "Appointment_ID":        f"APT{i:04d}",  # APT0001, APT0002 etc
        "Appointment_Date":      appt_date.strftime("%Y-%m-%d"),
        "Department":            dept,
        "Provider":              provider,
        "Appointment_Type":      appt_type,
        "Scheduled_Hour":        sched_time,
        "Slot_Duration_Mins":    slot_mins,
        "Wait_Time_Mins":        wait_time,
        "Status":                status,
        "Billing_Amount":        billing_amt if status == "Completed" else 0,
        "Claim_Status":          claim_stat,
        "Claim_Turnaround_Days": claim_days,
    })

# Convert list of rows into a pandas DataFrame (like an Excel table)
df = pd.DataFrame(rows)

# Save it as a CSV file
df.to_csv("appointments.csv", index=False)

print(f"✅ Dataset created! Total rows: {len(df)}")
print(f"\nFirst 3 rows preview:")
print(df.head(3))
```
